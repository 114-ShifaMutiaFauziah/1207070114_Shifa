```python
#import library

import matplotlib.pyplot as plt

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import numpy as np
from PIL import ImageEnhance
import numpy as np
import cv2
```

```python
import cv2
#percobaan 1 : crop image

#load image dari file 
ZootopiaImage= cv2.imread("C:/Users/Asus/Downloads/PRAKTIKUM 5_1207070114_Shifa Mutia Fauziah/Zootopia.jpg")
#mengubah gambar menjadi RGB
ZootopiaImage = cv2.cvtColor(ZootopiaImage, cv2.COLOR_BGR2RGB)

SabanaImage = cv2.imread("C:/Users/Asus/Downloads/PRAKTIKUM 5_1207070114_Shifa Mutia Fauziah/Sabana.jpg")
SabanaImage = cv2.cvtColor(SabanaImage, cv2.COLOR_BGR2RGB)

#menyimpan citra yang sudah di potong
ZootopiaCropped = ZootopiaImage.copy()
ZootopiaCropped = ZootopiaImage[0:256,64:320]

SabanaCropped = SabanaImage.copy().copy()
SabanaCropped = SabanaCropped[64:256,128:320]

#menampilkan citra asli dan hasil crop dalam 1 figure
fig, axes = plt.subplots(2, 2, figsize=(6, 6))
ax = axes.ravel()
ax[0].imshow(ZootopiaImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(SabanaImage, cmap='gray')
ax[1].set_title("Citra Input 2")
ax[2].imshow(ZootopiaCropped)
ax[2].set_title("Citra Output 1")
ax[3].imshow(SabanaCropped, cmap='gray')
ax[3].set_title("Citra Output 2")
```

```python
#Percobaan 2 - Citra Negative

#membalikan citra yang tersimpan
inv  = invert( ZootopiaCropped ) 
print ( 'Shape Input : ' ,  ZootopiaCropped . shape ) 
print ( ' Shape Output : ' , inv . shape )
#menampilkan gambar asli dan hasil invert beserta histogramnya dalam 1 figure
fig ,axes  =  plt . subplots ( 2 ,  2 ,  figsize = ( 6 ,  6 )) 
ax  =  axes .ravel ()
ax[0].imshow ( ZootopiaCropped )
ax[0].set_title ( "Input Citra" )
#menampilkan histogram gambar asli
ax[1].hist ( ZootopiaCropped .ravel ( ), bins = 256 ) 
ax[1].set_title ( 'Masukan histogram' ) 
ax[2].imshow ( inv ) 
ax[2].set_title ( 'Citra Output (Gambar Terbalik)' )
#menampilkan histogram hasil invert
ax[3].hist ( inv . ravel (),  bins = 256 ) 
ax[3].set_title ( 'Keluaran histogram' )
```

```python
# Percobaan 3 Increase Brightness

#membuat brightness pada output 
#mengubah tipe data menjadi float
copySabana = SabanaCropped.copy().astype(float)

#menambahkan parameter 100 pada setiap elemen aray copyCamera untuk brightness
for i in range (copySabana.shape[0]):
    for j in range (copySabana.shape[1]):
       copySabana[i,j] += 100
copySabana[copySabana > 255] = 255
copySabana = copySabana.astype(np.uint8)

#menampilkan gambar asli dan hasil brightness berikut histogramnya dalam 1 figure
fig, axes = plt.subplots(2,2, figsize=(6,6))
ax = axes.ravel()
ax[0].imshow(SabanaCropped, cmap ='gray')
ax[0].set_title("Citra Input")
ax[1].hist(SabanaCropped.ravel(), bins=256)
ax[1].set_title("Histogram Input")
ax[2].imshow(copySabana, cmap ='gray')
ax[2].set_title("Citra Output (Brightness)")
ax[3].hist(copySabana.ravel(), bins=256)
ax[3].set_title("Histogram Input")
```
